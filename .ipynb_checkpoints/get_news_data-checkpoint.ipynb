```python
import pandas as pd
```

```python
df = pd.read_csv('news.csv')
```

```python
for i in range(len(df.content)) :
    df.iloc[i].content = df.iloc[i].content.replace('[','')
    df.iloc[i].content = df.iloc[i].content.replace(']','')
    df.iloc[i].content = df.iloc[i].content.strip('\n')
    df.iloc[i].content = df.iloc[i].content.strip('/Reuters=')
```
